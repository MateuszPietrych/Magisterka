```python
import re

def clean_file(input_path, output_path):
    with open(input_path, 'r', encoding='utf-8') as file:
        content = file.read()

    # Removing all segments that start with "Type: " and end with "--------------------------------------------------------------------------------"
    cleaned_content = re.sub(r'Type: .*?--------------------------------------------------------------------------------', '', content, flags=re.DOTALL)

    with open(output_path, 'w', encoding='utf-8') as file:
        file.write(cleaned_content)

input_path = 'combined_output.txt'
output_path = 'cleaned2_output.txt'
clean_file(input_path, output_path)

print(f"Cleaned file saved as: {output_path}")


# then remove TV/Film and quests 
# remove "Do you really wish to know?" â€” Spoilers from the books and/or adaptations to follow!
#
# Remove:
#--------------------------------------------------------------------------------
#
#--------------------------------------------------------------------------------
#
#  regular expresion to ->  --------------------------------------------------------------------------------\n\s*".*
#
```

```python
import os

def split_and_save(input_path, output_folder):
    # Ensure the output folder exists
    os.makedirs(output_folder, exist_ok=True)

    with open(input_path, 'r', encoding='utf-8') as file:
        content = file.read()

    # Split the content by the delimiter
    sections = content.split("--------------------------------------------------------------------------------")

    for idx, section in enumerate(sections):
        # Trim whitespace from each section
        cleaned_section = section.strip()
        
        # Skip empty sections
        if not cleaned_section:
            continue

        # TODO: Perform any desired operation on the cleaned_section here
        # For demonstration, let's just save the section as-is

        if should_have_file(cleaned_section):
            # Save each section as a separate file
            fileText = get_id(cleaned_section) + "<;>"+ cleaned_section
            output_file = os.path.join(output_folder, f'section_{idx+1}.txt')
            with open(output_file, 'w', encoding='utf-8') as output:
                output.write(fileText)

            print(f"Saved: {output_file}")
        

def should_have_file(section):
    if (section.contains(",") or section.contains(".") or section.contains("is") or section.contains("are")  or section.contains("was") or section.contains("were")) and len(section) > MIN_LENGTH_OF_SECTION:
        return True
    return False

def get_id(section):
    # Extract the ID from the section using regex
    dotIndex = section.find(".")
    commaIndex = section.find(",")
    isIndex = section.find("is")
    areIndex = section.find("are")
    wasIndex = section.find("was")
    wereIndex = section.find("were")

    indexList  = [dotIndex, commaIndex, isIndex, areIndex, wasIndex, wereIndex]
    indexList = [index for index in indexList if index != -1]

    if indexList:
        minIndex = min(indexList)
        return section[:minIndex].strip()
    return section.strip()

# Define the input file and output folder
input_path = 'cleaned_output.txt'
output_folder = 'split_files'
MIN_LENGTH_OF_SECTION = 30

# Call the function
split_and_save(input_path, output_folder)
```
